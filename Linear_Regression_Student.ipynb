```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=df.dropna() #removes missing rows
```

```python
X= df[["hours_studied","sleep_hours","attendance_percent","previous_scores"]]
y=df["exam_score"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
#Linear Regression
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

print(mean_squared_error(y_test, y_pred))
print(r2_score(y_test, y_pred))
```

```python
#Visualization on Predicted and Actual values.
sns.scatterplot(x=y_test, y=y_pred)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.show()
```

```python
#Comparison
sns.regplot(x='hours_studied', y='exam_score', data=df)
plt.title("Regression Line: Study Hours vs Score")
plt.show()
```
