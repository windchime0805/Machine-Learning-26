```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
# Load dataset
df = pd.read_csv("/content/student_exam_scores.csv")
```

```python
# Drop ID column
df = df.drop('student_id', axis=1)

#Dropping this because in our table it is in string format which raises error
#while training the model
```

```python
# Features and target
X = df.drop('exam_score', axis=1)
#Considering only factors(so we drop the result) axis=1 -> dropping the column.
y = df['exam_score']
# Considering only result i.e.., "exam_score".
```

```python
# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5)
```

```python
# Train model
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Predict
y_pred = model.predict(X_test)
```

```python
# Evaluate
print("R2 Score:", r2_score(y_test, y_pred))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(x=y_test, y=y_pred)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Actual vs Predicted Scores")
plt.show()
```
