```python
import pandas as pd
```

```python
# Load dataset
df = pd.read_csv("/content/s_algo.csv")
```

```python
# Clean target column
df['EnjoySport'] = df['EnjoySport'].str.strip().str.capitalize()

X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
# Initialize S and G
S = X[0].copy()
G = [["?" for _ in range(len(S))]]

for i in range(len(X)):
    if y[i] == "Yes":
        for j in range(len(S)):
            if S[j] != X[i][j]:
                S[j] = "?"
                G[0][j] = "?"
    else:
        for j in range(len(S)):
            if S[j] != X[i][j]:
                G[0][j] = S[j]
            else:
                G[0][j] = "?"

print("Specific Hypothesis (S):", S)
print("General Hypothesis (G):", G)
```
