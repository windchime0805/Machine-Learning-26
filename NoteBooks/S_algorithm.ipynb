```python
import pandas as pd

# Load dataset
df=pd.read_csv("/content/s_algo.csv")

# Clean target column
df['EnjoySport'] = df['EnjoySport'].str.strip().str.capitalize()

# Split features and target
X = df.iloc[:, :-1]
y = df.iloc[:, -1]

# Initialize hypothesis with first positive example
positive_rows = X[y == "Yes"]

hypothesis = positive_rows.iloc[0].tolist()

# Apply Find-S algorithm
for i in range(len(X)):
    if y[i] == "Yes":
        for j in range(len(hypothesis)):
            if hypothesis[j] != X.iloc[i, j]:
                hypothesis[j] = "?"

print("Final Hypothesis:", hypothesis)
```
